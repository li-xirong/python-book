# 使用Matpotlib绘制正弦曲线 
正弦曲线函数形式：$y=sin(x)$

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'Heiti TC', 'Hiragino Sans GB', 'PingFang SC'] # 正常显示中文
plt.rcParams['axes.unicode_minus'] = False # 正常显示负号
# 添加魔法命令，让图表直接显示在Notebook代码单元格下方
%matplotlib inline
```

```python
fig, axes = plt.subplots() # 创建画布和子图
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
axes.plot(x, y) # 调用子图的绘图功能
plt.savefig('sine.png') # 将绘图结果保存为本地图片文件
```
