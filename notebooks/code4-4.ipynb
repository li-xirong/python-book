# 使用BeautifulSoup和正则表达式解析网页标题

```python
import re
from bs4 import BeautifulSoup
```

```python
pattern = r"^(\d{4})\D(.+)(省|市|自治区)"
html_file = 'html_pages/aHR0cHM6Ly9yZHpzLnJ1Yy5lZHUuY24vaW5xdWlyeS9hZG1pc3Npb24vaW5kZXhjbXMv5YyX5LqsLzIwMjMvbGlzdGNtcw.html'
html_str = open(html_file).read() 
soup = BeautifulSoup(html_str, 'html.parser')
tag = soup.find('h1', class_='y_tit')
title = tag.get_text().strip()
m = re.search(pattern, title)
year, pro, pro_type = m.group(1, 2, 3)
print(title)
print(year, pro, pro_type)
```
