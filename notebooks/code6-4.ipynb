# 绘制历年人大工商管理专业在福建省理科录取分数线折线图（改进版）
添加坐标轴标签、样本点数值、增大刻度字体、设置网格线等

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'Heiti TC', 'Hiragino Sans GB', 'PingFang SC'] # 正常显示中文
plt.rcParams['axes.unicode_minus'] = False # 正常显示负号
```

```python
fig, axes = plt.subplots()
years = list(range(2015, 2024))
scores = [671, 645, 608, 664, 650, 673, 664, 658, 671] 
axes.plot(years, scores, '-o')
# 为每个样本点添加分数标识
for year, score in zip(years, scores): 
    axes.annotate(score, xy=(year, score), ha='center', va='bottom', fontsize=12)
# 设置坐标轴标签
axes.set_xlabel("年份", fontsize=14) #, labelpad=8)  # 设置x轴标签
axes.set_ylabel("理科最低录取分数", fontsize=14) #, labelpad=8)  # 设置y轴标签
# 设置坐标轴显示范围
axes.set_xlim(2014.5, 2023.5)
axes.set_ylim(600, 700)
# 设置x坐标轴刻度
axes.set_xticks(years) 
# 设置横纵坐标轴刻度字体大小
axes.tick_params(axis='x', labelsize=14) 
axes.tick_params(axis='y', labelsize=14)
# 设置网格线
axes.grid(linestyle='-.', linewidth=0.5)
```
