# 列表对象、集合对象的创建与简单操作。

```python
lst = [] #创建一个名为lst的空列表对象
lst.append(1) # 往列表中添加一个整数1
print(len(lst)) # 通过len函数获得lst中的元素个数（即1）并通过print函数输出到屏幕上
lst.append(2) # 往列表中再添加一个整数2
print(len(lst)) # 输出 2
print(id(lst)) # 输出lst所指列表对象的标识符，形如140535170940736

st = set() #创建一个名为st的空集合对象
st.add(1) #往集合中添加一个整数1
print(len(st)) # 获取st中的元素个数（即1）并输出到屏幕上
st.add(2) #往集合中再添加一个整数2
print(len(st)) # 输出 2
print(type(st)) # 输出 <class 'set'>
```
